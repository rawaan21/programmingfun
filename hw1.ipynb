```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/Users/rawaanakhtar/Desktop/whatweuse.csv')


```

```python
df.describe()
```

```python
df.plot(x='Date', y='Price', kind='line')
```

```python
df['Price'].rolling(window=12).mean().plot()
```

```python
N = df.shape[0]
training_size = int(N*0.6)

df_train, df_test = train_test_split(df, train_size=training_size, random_state=0)

X_train = df_train.iloc[:, :-1]
y_train = df_train.iloc[:, -1]

X_test = df_test.iloc[:, :-1]
y_test = df_test.iloc[:, -1]

model = LinearRegression().fit(X_train, y_train)

predictions = model.predict(X_test)
print(predictions)
```
